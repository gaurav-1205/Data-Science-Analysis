```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import boxcox
```

```python
#Read the dataset and assign values of e to a varible
dataset = pd.read_csv('exoplanet.eu_catalog.csv')
eccentricity = dataset['eccentricity']
eccentricity_withoutNa = eccentricity.dropna()     #We drop the NaN rows from the eccentricity column
```

```python
#plot the histogram of 'e'
plt.hist(eccentricity_withoutNa, bins=15)
plt.title('Eccentricity Histogram')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#plot the gaussian distribution using boxcox
posdata = eccentricity[eccentricity>0]
e_boxcox, const = boxcox(posdata)
plt.hist(e_boxcox, bins=15)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Eccentricity Gaussian Distribution using boxcox transformation')
plt.show()
```
